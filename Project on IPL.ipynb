```python
# in our dataset there is total runs column so we did group by on following two columns to fetch total runs of the team

total_score = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score
```

```python
# HERE WE ONLY FETCH 1ST INNINGS DATA BECAUSE ON THAT SCORE WE WILL PREDICT THE WIN

total_score = total_score[total_score['inning']==1]
```

```python
# HERE WE WILL MERGE THE DATA OF MATCHES WITH TOTAL SCORE ON BELOW COLUMNS

match_df = matches.merge(total_score[['match_id','total_runs']], left_on='id',right_on='match_id')
match_df
```

```python
# UNIQUE TEAMS IN OUR DATASET

match_df['team1'].unique()
```

```python
# TEAMS THAT WE REQUIRED 

teams = ['Sunrisers Hyderabad','Mumbai Indians',
        'Royal Challengers Bangalore','Kolkata Knight Riders',
        'Kings XI Punjab','Chennai Super Kings','Rajasthan Royals',
         'Delhi Capitals']
```

```python
# WE WILL REPLACE SOME TEAMS NAMES

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals') 
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals') 

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
# WE WANT ONLY THOSE MATCHES WHERE DL IS NOT APPLIED 

match_df['dl_applied'].value_counts()
```

```python
# NOW WE WILL TAKE ONLY THOSE DATA WHERE DL ==0

match_df = match_df[match_df['dl_applied']==0]
match_df
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delevery_df = match_df.merge(deliveries,on='match_id')
delevery_df
```

```python
delevery_df = delevery_df[delevery_df['inning']==2]
delevery_df
```

```python
delevery_df.info()
```

```python
#delevery_df.loc[:, 'total_runs_y'] = pd.to_numeric(delevery_df['total_runs_y'])
delevery_df['total_runs_y'] = pd.to_numeric(delevery_df['total_runs_y'], errors='coerce')
```

```python
# Convert column to numeric, coercing non-numeric values to NaN
delevery_df['total_runs_y'] = pd.to_numeric(delevery_df['total_runs_y'], errors='coerce')

# Check for NaN values in the column
nan_values = delevery_df['total_runs_y'].isna().sum()
if nan_values > 0:
    print(f"Warning: Found {nan_values} non-numeric values in 'total_runs_y' column.")

# Handle NaN values as needed (e.g., remove rows with NaN values or impute missing values)

# Retry the cumulative sum operation
cumulative_sum = delevery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delevery_df['current_score'] = delevery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delevery_df
```

```python
delevery_df['runs_left'] = delevery_df['total_runs_x'] - delevery_df['current_score']
```

```python
delevery_df.head()
```

```python
delevery_df['balls_left'] = 126- (delevery_df['over']*6 + delevery_df['ball'])
```

```python
delevery_df.sample(2)
```

```python
delevery_df['player_dismissed'] = delevery_df['player_dismissed'].fillna('0')
delevery_df['player_dismissed'] = delevery_df['player_dismissed'].apply(lambda x: x if x=='0' else '1')
delevery_df['player_dismissed'] = delevery_df['player_dismissed'].astype('int')
```

```python
wickets = delevery_df.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
delevery_df['wickets'] = 10- wickets
```

```python
delevery_df.tail()
```

```python
# crr = runs/overs

delevery_df['crr'] = (delevery_df['current_score']*6)/(120- delevery_df['balls_left'])
```

```python
# rrr
delevery_df['rrr'] = (delevery_df['runs_left']*6/ delevery_df['balls_left'])
```

```python
delevery_df.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delevery_df['result'] = delevery_df.apply(result, axis=1)
```

```python
final_df = delevery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df.sample(3)
```

```python
# we will shuffle the rows to avoid biasness

final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
# IN X_TRAIN FOR rrr COLUMN MIN AND MAX VALUES ARE COMING INFINITY SO WE WILL DELETE THAT ROWS WHERE BALLS==0
final_df = final_df[final_df['balls_left'] != 0]
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.22, random_state=36)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False, drop='first'),['batting_team','bowling_team','city'])
], remainder= 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
x_train.describe()
```

```python
y_pred= pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```
